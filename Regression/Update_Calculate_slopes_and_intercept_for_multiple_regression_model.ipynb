***Model***

```python
import pandas as pd
AA1 = pd.read_csv('Example2.csv', header=0)
AA1.head()
```

```python
TrainX = AA1.drop(['T'], axis='columns')
TrainX.head()
```

```python
TrainY=AA1['T']
TrainY
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(TrainX, TrainY)
```

```python
# Get the slope (coefficient) and intercept
slope = model.coef_
intercept = model.intercept_

print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
```

```python
model.predict([[3,7,4,17,12,24]])
```
